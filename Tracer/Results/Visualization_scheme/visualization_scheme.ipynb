```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import scipy
```

```python
DATA_FILE = 'C:/_Repositories/KaschenkoEquation/Theory/Scheme/scheme.csv'
SAVE_FILE = 'C:/_Repositories/KaschenkoEquation/Theory/Scheme/scheme.png'
```

```python
data = pd.read_csv(DATA_FILE, sep = ',')
data.head()
```

```python
alpha_cr = data[['g', 'a']]
#alpha_cr = alpha_cr.dropna(how='any')
alpha_cr.head()
```

```python
beta = data[['g', 'b']]
beta.head()
```

```python
mpl.rcParams['font.size'] = 26
mpl.rc('lines', linewidth=5)
plt.figure(figsize=(22, 12))
plt.xlabel('γ')
plt.ylabel('α')
plt.minorticks_on()
plt.grid(True)
Xa = alpha_cr['g']
Xb = beta['g']
A = alpha_cr['a']
B = beta['b']
plt.plot(Xa, A, label='α_c', color='orangered', zorder=2)
plt.plot(Xb, B, label='α_u', color='steelblue', zorder=2)
plt.fill_between(Xa, min(A), A, facecolor='bisque', interpolate=True)
plt.fill_between(Xa, A, B, facecolor='palegreen', interpolate=True)
plt.text(3.3, -4.6, r'$\alpha_c$', fontsize=28)
plt.text(1.2, -6.5, r'$S$', fontsize=48)
plt.text(4.5, -1.5, r'$\alpha_u$', fontsize=28)
plt.text(2.8, -1.3, r'$U$', fontsize=48)
plt.text(2.8, -12.3, r'$C$', fontsize=48)
plt.axhline(y=-1.825, xmin=0.045, xmax=0.84, color='grey', ls='--', linewidth=3, zorder=1)
plt.axvline(x=4.1, ymin=0.05, ymax=0.86, color='grey', ls='--', linewidth=3, zorder=1)
plt.scatter(4.1, -1.825, marker='o', color='black', s=98, zorder=3)
rect = patches.Rectangle((0.0, -17.5), 4.1, 17.5, linewidth=1, edgecolor='black', facecolor='none')
#plt.legend()
plt.savefig(SAVE_FILE)
```
